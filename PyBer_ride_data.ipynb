```python
#import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
#Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
#Plot the line chart
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
#Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

#Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", figsize=(7,7))
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
#Plot graph
y=pyber_ride_df["Avg. Fare ($USD)"]
tick_locations = [value for value in x_axis]
stdev = statistics.stdev(y)

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=stdev, capsize=3, color="skyblue")

#Rotate labels
plt.xticks(tick_locations, pyber_ride_df["Month"], rotation=0)
plt.yticks(np.arange(0, 61, step=5.0))

plt.show()
```
